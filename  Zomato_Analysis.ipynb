# Zomato Data Analysis 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\tanuj\\Downloads\\Zomato data .csv")
df
```

```python
df['name'].max()
```

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)
df['rate'] = df['rate'].apply(handleRate)
print(df.head())
```

```python
df.info()
```

# What type of restaurarent do the majority of customer order from ?

```python
df.head()
```

```python
sns.countplot(x = df['listed_in(type)'], palette = 'rainbow')
plt.xlabel('Type of restaurarent')
```

# How Many Votes has Each type of res. received from customer ?

```python
df.head()
```

```python
group = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': group})
plt.plot(result, color = 'green', marker = 'o')
plt.xlabel('Type of Restarurant', c = 'blue', size = 20)
plt.ylabel('votes', c = 'red', size = 20)
```

# Conclusion
## Dining has maximum Votes

# Which type of restaurant received more offline orders , so that Zomato can  provide customers with  some good offers ?

```python
pivot_table = df.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table, annot = True, cmap = 'YlGnBu', fmt = 'd')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed in (type)')
plt.show()
```

# Conclusion
## Dining Res primarly accept offline orders whear as cafes primarily receive online orders. This suggest that clients prefers orders in persom at restaurants ,  but prefer online ordering at cafes.
